```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Chennai_Central.jpg')  
plt.imshow(image[:,:,::-1])
plt.title('Original Image')
plt.axis('off')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

1. Sobel Edge Detection

```python
sobelx  = cv2.Sobel(src = gray_image, ddepth = cv2.CV_64F, dx = 1, dy = 0, ksize = 3) 
sobely  = cv2.Sobel(src = gray_image, ddepth = cv2.CV_64F, dx = 0, dy = 1, ksize = 3)

sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)  
sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)  
sobel_combined = cv2.magnitude(sobelx, sobely) 
```

```python
plt.figure(figsize = (12,16))

plt.subplot(321)
plt.axis('off')
plt.imshow(image[:,:,::-1])
plt.title('Original')

plt.subplot(322)
plt.axis('off')
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale')

plt.subplot(323)
plt.axis('off')
plt.imshow(sobelx)
plt.title('Sobel-X Edge Map')

plt.subplot(324)
plt.axis('off')
plt.imshow(sobely)
plt.title('Sobel-Y Edge Map')
```

```python
plt.figure(figsize = (7,7))

plt.axis('off')
plt.imshow(sobel_combined, cmap='gray')
plt.title('sobel_combined')

plt.show()
```

2.Laplacian Edge Detector

```python
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
```

```python
plt.figure(figsize = (12,16))

plt.subplot(121)
plt.axis('off')
plt.imshow(gray_image, cmap='gray')
plt.title('Inputimage (Gray Image)')

plt.subplot(122)
plt.imshow(laplacian, cmap='gray')
plt.axis('off')
plt.title('Output Image (laplacian)')

plt.show()
```

3. Canny Edge Detection 

```python
edges = cv2.Canny(gray_image, threshold1 = 180, threshold2 = 200)
```

```python
plt.figure(figsize = (12,16))

plt.subplot(221)
plt.axis("off")
plt.imshow(image[:,:,::-1])
plt.title('Original')

plt.subplot(222)
plt.axis("off")
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale')
```

```python
plt.figure(figsize = (7,7))

plt.axis("off")
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Map')

plt.show()
```

4. Prewitt Edge Detection

```python
prewitt_x = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
prewitt_y = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
prewitt_x_edge = cv2.filter2D(gray_image, -1, prewitt_x)
prewitt_y_edge = cv2.filter2D(gray_image, -1, prewitt_y)
prewitt = cv2.magnitude(prewitt_x_edge.astype(np.float32),prewitt_y_edge.astype(np.float32))
```

```python
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.imshow(prewitt, cmap='gray')
plt.title("Prewitt Edge Detection")
plt.axis("off")

plt.show()
```

5. Roberts Edge Detection

```python
roberts_x = np.array([[1,0],[0,-1]])
roberts_y = np.array([[0,1],[-1,0]])
roberts_x_edge = cv2.filter2D(gray_image, -1, roberts_x)
roberts_y_edge = cv2.filter2D(gray_image, -1, roberts_y)
roberts = cv2.magnitude(roberts_x_edge.astype(np.float32),roberts_y_edge.astype(np.float32))
```

```python
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.imshow(roberts, cmap='gray')
plt.title("Roberts Edge Detection")
plt.axis("off")

plt.show()
```
